30. Version management:

To change the default version:


1. nvm alias default 12.13.0

2. nvm use 12.13.0

3. npm get prefix

4. change accordingly in .bash_profile

----
Set and unset the version:

1. npm config set prefix /Users/sreedhar.k/.nvm/versions/node/v12.13.0/lib

2. nvm use --delete-prefix v12.13.0 --silent

----
To check for installed versions of the package:
1. npm view <package_name> version

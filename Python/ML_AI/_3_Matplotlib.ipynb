### Subplots and legends

Matplotlib is strongly object oriented and its principal objects are the figure and the axes.

You can think of the figure as a canvas, of which you typically specify the dimensions and possibly e.g., the background color etc etc. You use the canvas, the figure, essentially in two ways, placing other objects on it (mostly axes, but also text labels etc) and saving its contents with savefig.

You can think of an axes as a sort of Swiss Army knife, a handy object that offers a tool (e.g. .plot, .scatter, .hist etc) for everything, mostly. You can place one, two, ... many axes inside a figure using one of many different methods.

```python
x = np.arange(0,10)
y = 2*x
z = x**3

def first():
    fig, axes = plt.subplots(nrows=1,ncols=2)
    display(
        type(axes),
        type(fig),
    )

    display(
        axes[1].plot(x,y),
        axes[0].plot(z,y),
        axes[1].plot(z,x),
        axes[0].plot(y,x),
    )
#first()


def second():
    fig, axes = plt.subplots(nrows=2,ncols=2)

    display(
        axes[1][0].plot(x,y),axes[1][0].set_ylabel("Label_10"),axes[1][0].set_title("Title_10"),
        axes[0][0].plot(z,y),axes[0][0].set_ylabel("Label_00"),axes[0][0].set_title("Title_00"),
        axes[1][0].plot(z,x),axes[1][0].set_ylabel("Label_10"),axes[1][0].set_title("Title_10"),
        axes[0][1].plot(y,x),axes[0][1].set_ylabel("Label_01"),axes[0][1].set_title("Title_01"),
        fig.subplots_adjust(wspace=0.9,hspace=0.5),
        fig.suptitle('2x2 graphs'),
    )
    fig.savefig('images/new_grid_figure.png',bbox_inches = 'tight')
#second()

def legends():
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.plot(x,x,label='X vs X')
    ax.plot(x,x**2,label = 'X vs X**2')
    ax.legend()
#legends()

def styles():
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.plot(
        x,x,label='X vs X',
        color='purple',
        marker="*",
        markersize=10,
    )
    ax.plot(
        x,x**2,label = 'X vs X**2',
        color='black',
        linewidth=0.5,
        linestyle=":",
        marker="o",
        markersize=10,
    )
    ax.plot(
        x,x**3,label = 'X vs X*3',
        color='red',
        linewidth=1.5,
        linestyle="-.",
        marker="+",
        markersize=20,
    )
    ax.legend()   
styles()


hide_toggle()
```

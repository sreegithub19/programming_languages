def isprime(self,m):
        if(m==2 or m==3 or m==5):return True
        if(m==1 or m==4):return False
        for i in range(2,(m//2)):
            if(m%i==0):
                return False
        return True


2. Python binary tree template code:

class Solution:
    def traverse(self,root,li):
        if(root is not None):
            li.append(root.val)
            self.traverse(root.left,li)
            self.traverse(root.right,li)
        return li
    def findMode(self, root: Optional[TreeNode]) -> List[int]:
        li = []
        ans = self.traverse(root,li)
        count = []
        for i in ans:
            count.append(ans.count(i))
        return sorted(set([ans[i] for i in range(len(ans)) if(count[i]==max(count))]))

4. Python template for linked list:

class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if(head is None):
            return None
        li = []
        curr = head
        while curr:
            li.append(curr.val)
            curr = curr.next
        li = [i for i in li if(li.count(i)==1)]
           
        previous = None
        first = None
        for i in li:
            current = ListNode(i)
            if(first is None):
                first = current
            if(previous is not None):
                previous.next = current
            previous = current
        return first

class Solution:
    def compress(self, chars: List[str]) -> int:
        #print(chars)
        char = "".join(sorted(set(chars)))
        ans = ""
        for i in range(len(char)):
            ans += char[i]
            if(chars.count(char[i]) > 1):
                ans += str(chars.count(char[i]))
        #print(ans)
        ans = [i for i in ans]
        #print(ans)
        chars[:] = ans   #in-place replacement

Brute force:
    m = target sum, n = array length
    Space complexity: O(m)
    Time complexity: O(n**m)
Memoized:
    m = target sum, n = array length
    Space complexity: O(m)
    Time complexity: O(n*m)
'''
def can_sum(target, numbers,memo):
    if(target in memo):
        return memo[target]
    if(target==0):
        return True
    if(target <0):return False
    
    for num in numbers:
        remainder = target - num
        #print(remainder)
        if(can_sum(remainder,numbers,memo)==True):
            memo[target] = True
            return True
    memo[target] = False
    return False

class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        p = ''.join(sorted(p))
        ans = []
        
        for i in range(len(s) - len(p) + 1):
            flag=0
            for val in "".join([chr(j) for j in range(97,123)]):

            if(flag==0):

class Solution:
    def addNegabinary(self, arr1, arr2):
        n = (sum(pow(-2, i) for i, v in enumerate(arr1[::-1]) if v) +
             sum(pow(-2, i) for i, v in enumerate(arr2[::-1]) if v))
        res = [] if n else [0]
        while n:
            n, rem = divmod(n, -2)
            if rem < 0:
                n += 1
            res.append(abs(rem))
        res.reverse()
        return res
